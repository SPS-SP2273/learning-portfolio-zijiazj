# 1 A simple plot

## 1.1 Letâ€™s look at some code.

````python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values',
         color='red', linestyle='dashed')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
````

### Things to note

| Long form | Abbreviation |
| :--------: | :----------: |
| color     | c            |
| linestyle | ls           |
| linewidth | lw           |

```python
# Both will produce the same result
plt.plot(x, y, color='red', linestyle='dashed', linewidth=2)
plt.plot(x, y, c='red', ls='dashed', lw=2)
```

plotting functions usually have default values for the styling parameters.

````python
plt.plot(x, y_1, y_2)
````

````python
plt.plot(x, y_1, color='red', linestyle='dashed', linewidth=2)
````

````python
# the order you specify keywords argument does not matter
# color, lw, ls, label
plt.plot(x, y_1, label='Y values',
           color='red', linestyle='dashed')
````

## 1.2 Adding another plot

````python
plt.plot(x, y_2,
         label='Y2 values', color='blue')
plt.plot(x, y_1, label='Y values',
           color='red', linestyle='dashed')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
````

## 1.3 Yet another plot but with error bars

Error bars are lines extending from each data point along the y-axis (or x-axis if horizontal). These bars represent the uncertainty or variability associated with each data point.

Syntax of errorbar: plt.errorbar(x, y, xerr=x_error, yerr=y_error)

````python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             label='Y3 with errors', color='green')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
````

# 2 Better with NumPy

advantages of NumPy arrays is that they allow us to generate data-related mathematical functions easily. 

````python
import numpy as np
````

````python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')

plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
````

## 2.1 Adding mathematical functions

````python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                  # The math stuff
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()                                 
````

our plot does not look good because sin(x) lies between +/- 1, but x^2 has no such bounds. 

One way to fix this is to add another y-axis that shares the same x-axis.

## We need another axis!

use the command twinx()

````python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
````

### Things to note

- We now have two legend() calls, one for each axis.
- 
our plot still does not look good because we have only a few points. Letâ€™s use np.linspace to fix this

````python
x = np.linspace(0, 6, 1000) #np.linspace(start, stop, step)
x2 = x**2
sin_x = np.sin(x)


plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1
plt.ylabel('y1-values')

# anything above plt.twinx is the left axis
plt.twinx()                              
# anything below plt.twinx is the right axis

plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
plt.ylabel('y2-values')
````

# 3 Saving to disc

- export plots into many formats (PDF, JPEG, PNG, BMP, etc.) and at different resolutions.
- We need to use the function savefig() for this.
- We specify the format with the extension (e.g., filename.pdf) of the file name,
- We specify resolution by use dots-per-inch (dpi).

````python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x, sin_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')

# saving the figure to the same directory
plt.savefig('simple-plot.png', dpi=150)
````

# 4 A real example: Global Warming

## 4.1 Plotting data from files

````python
# numpy will read the txt file and store them in 2D array
# ignore the first 24 lines in the text file (description of the data)
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)

#the first element of all subarray is date
date = data[:, 0]
#the last element of all subarray is anomaly
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=0.5)
plt.ylim([-8, 8])


# alpha parameter determines the transparency of the plotted line 
# floating-point value between 0 and 1
# 0 is completely transparent (invisible) and 1 is completely opaque (fully visible).
````

````python
print(data)
````

add a horizontal line at the zero value to highlight the trend shown by the data.

````python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=0.5)
plt.ylim([-8, 8])

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
````

## 4.2 Styles

Matplotlib offers some [standard style templates](https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html) we can directly use

````python
# this function must be placed in the first row of our codes
plt.style.use('fivethirtyeight')

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
````
